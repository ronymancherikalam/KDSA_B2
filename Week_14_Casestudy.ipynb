# Unsupervised Learning

### Importing models and metrics

```python
from sklearn.cluster import KMeans
import scipy.cluster.hierarchy as sch
from scipy.cluster.hierarchy import dendrogram,linkage
from sklearn.cluster import AgglomerativeClustering
from kneed import KneeLocator
from sklearn.metrics import silhouette_score
```

# Kmeans Clustering

```python
wcss=[]
silhouette_coefficients = []
for i in range(2,11):
    kmeans=KMeans(n_clusters=i,init='k-means++',random_state=32)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
    score = silhouette_score(X, kmeans.labels_)
    silhouette_coefficients.append(score)
    print('WCSS-Cluster',i,'=',kmeans.inertia_)
```

### Elbow Method

```python
plt.plot(range(2,11),wcss,'-x')
plt.title('Elbow Method')
plt.xlabel('No of Clusters')
plt.ylabel('WCSS/SSE')
plt.show()
```

##### As per Elbow method no of cluster should be 3

### Kneelocator Method

```python
knee=KneeLocator(range(2,11),wcss,curve='convex',direction='decreasing')
knee.elbow
```

### Validatiing no of clusters using silhouette_coefficient
We have already calculated score while doing kmeans cluster identifiaction previously. plotting the graph

```python
plt.plot(range(2, 11), silhouette_coefficients,'-x')
plt.xlabel("Number of Clusters")
plt.ylabel("Silhouette Coefficient")
plt.show()
```

##### Silhoute Coefficient is giving cluster no 3 having highest score. so we will chose 3 as no of clusters

### Fitting Kmeans with no of cluster as 3

```python
kmeans=KMeans(n_clusters=3,init='k-means++',random_state=32)
y_kmeans=kmeans.fit(X)
cluster=kmeans.labels_
cluster
```

### Mapping the clusters to observations in Data Frame

# Agglomerative Heirarchical Clustering

```python
#Dendogram
plt.figure(figsize=(8,6))
dgram=sch.dendrogram(sch.linkage(X,method='ward'))
plt.title('Dendogram')
plt.xlabel('Cluster')
plt.ylabel('Euclidean Distance')
plt.hlines(20,0,5000,colors='red',linestyles='dashed')
plt.show()
```
