```python
# So there are white spaces and we have to remove or correct. Recalculating Totalcharges column as below
```

```python
ten_chrg['TotalCharges']=ten_chrg['tenure']*ten_chrg['MonthlyCharges']
```

```python
# Checking white space again
```

```python
ten_chrg[ten_chrg['TotalCharges']==' '].sum().any()
```

```python
# Plotting graph after data cleaning
```

```python
plt.figure(figsize=(8,8))
plt.scatter(ten_chrg['tenure'],ten_chrg['TotalCharges'],marker='>')
plt.title('Comaprison of Tenure vs TotalCharges')
plt.xlabel('Tenure')
plt.ylabel('TotalCharges')
```

```python
# Inference: As the tenure is increasing total charges paid is proportionaly increasing as total charge is a 
# multiplier of tenure and MonthlyCharges and hence follow a linear relationship
```
