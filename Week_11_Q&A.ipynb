# Label Encoding


```python
#importing label encoder from skleran library
from sklearn.preprocessing import LabelEncoder
#creating a label encoder instance
Label_enc=LabelEncoder()
for i in data.select_dtypes(include=['object']).columns:
    if i!='passed':
        data[i]=Label_enc.fit_transform(data[i])
        continue
    
```

```python
data.head()
```

These columns are label encoded.

# one hot encoding 
school,sex,address,famsize,Pstatus,schoolsup,famsup,paid,activities,nursery,higher,internet,romantic columns


```python
data=pd.get_dummies(data=data,columns=['school','sex','address','famsize','Pstatus','schoolsup','famsup',
               'paid','activities','nursery','higher','internet','romantic'])
```

```python
data.head()
```

```python
data.info()
```

```python
# all categorical variables are converted to numerical.
```

# Splitting the data to train the model, Passed varibale is taken as the target variable

```python
X = data.drop(['passed'], axis=1)
y = data['passed']
#importing train_test_split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 42, test_size = 0.2)
```

```python
#Fitting the model SVM with Gaussian Kernel
#importing SVM
from sklearn.svm import SVC
```

```python
svm_gauss=SVC(kernel='rbf')
svm_gauss.fit(X_train,y_train)
y_pred=svm_gauss.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
print('Accuracy Score:',accuracy_score(y_test,y_pred))
print('Conclusion matrix\n',confusion_matrix(y_test,y_pred))
```

# Random Forest model fitting

```python
from sklearn.ensemble import RandomForestClassifier
rf_mod = RandomForestClassifier(random_state=42)
rf_mod.fit(X_train,y_train)
y_pred=rf_mod.predict(X_test)
print('Accuracy Score:',accuracy_score(y_test,y_pred))
print('Conclusion matrix\n',confusion_matrix(y_test,y_pred))
```

SVM model is giving an accuracy of 65.8% where as random forest is giving 72% accuracy

```python
# trying standard scaling
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.fit_transform(X_test)
```

```python
#Random forest model after standard scaling
rf_mod.fit(X_train,y_train)
y_pred=rf_mod.predict(X_test)
print('Accuracy Score:',accuracy_score(y_test,y_pred))
print('Condusion matrix\n',confusion_matrix(y_test,y_pred))
```

```python
#SVM model after standard scaling
svm_gauss.fit(X_train,y_train)
y_pred=svm_gauss.predict(X_test)
print('Accuracy Score:',accuracy_score(y_test,y_pred))
print('Conclusion matrix\n',confusion_matrix(y_test,y_pred))
```

After standard scaling we can see there is a slight improvement in accuray of both models. But random forest is giving better accuracy at 73.14%
