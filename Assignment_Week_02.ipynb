```python
# Visualisations
```

```python
sns.pairplot(data,hue="Classification")
```

```python
sns.heatmap(data.corr(),annot=True)
```

```python
# heat maps shows there is high correlation between PL-PW , 
#correlation is high for SL-PL and SL-PW. Regarding -ve correlation none of the variables 
#have high inverse relationship on other
```

```python
plt.figure(figsize=(15,7))
plt.subplot(1,2,1)
plt.scatter(x='SL',y='SW',data=data,color='violet')
plt.title('Sepal Comparison',fontsize=16)
plt.xlabel('Sepal Length')
plt.ylabel('sepal Width')
plt.legend(loc=1)
plt.subplot(1,2,2)
plt.scatter(x='PL',y='PW',data=data,color='red')
plt.title('Petal Comparison',fontsize=16)
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.tight_layout()
```

```python
#  As per above plot Some species petals are smaller and sepal distribution need deeper analysis
```

```python
sns.scatterplot(x='PL',y='PW',data=data,hue='Classification')
```

```python
# Comparitively iris-setosa has a smaller petal Size and largest being Iris-virginica and versicolour comes in the median
```

```python
sns.scatterplot(x='SL',y='SW',data=data,hue='Classification')
```

```python
# As per above plot we cannot deduce an impact on SL-SW properly and hence -ve correlation is close to 0
```

```python
sns.scatterplot(x='SL',y='PL',data=data,hue='Classification')
plt.xlabel('Sepal Length')
```

```python
# setosa has less length PL and SL since flower is smaller compared to other two
```

```python
sns.scatterplot(x='SL',y='PW',data=data,hue='Classification')
```

```python
# same deduction as above can be arrived in this also.  setosa is the smallest flower and viriginica is largest among irises
```

```python
plt.figure(figsize=(20,10))
plt.subplot(1,4,1)
sns.violinplot(x='Classification',y='PL',data=data)
plt.subplot(1,4,2)
sns.violinplot(x='Classification',y='SL',data=data)
plt.subplot(1,4,3)
sns.violinplot(x='Classification',y='PW',data=data)
plt.subplot(1,4,4)
sns.violinplot(x='Classification',y='SW',data=data)
plt.tight_layout
```

```python
# As per above violin chart we can cleary infer that setosa is the smallest among species though they have more width and
# Virginica is the largest and versicolor being the medium sized flower of species
```
