### Using Standard Scaling
##### Data points in columns are in different scales so using standard scaling to scale data uniformily

```python
df1= df.drop(['country'],axis=1)
cols= df1.columns
from sklearn.preprocessing import StandardScaler
st_scale= StandardScaler()
df1=st_scale.fit_transform(df1)
```

```python
X=pd.DataFrame(df1,columns=[cols])
X
```

##### All columns have been standard scaled

### K means Clustering

```python
from sklearn.cluster import KMeans
from kneed import KneeLocator
from sklearn.metrics import silhouette_score
```

```python
wcss=[]
silhouette_coefficients = []
for i in range(2,11):
    kmeans=KMeans(n_clusters=i,init='k-means++',random_state=32)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
    score = silhouette_score(X, kmeans.labels_)
    silhouette_coefficients.append(score)
    print('WCSS-Cluster',i,'=',kmeans.inertia_)
```

### Elbow Method

```python
plt.plot(range(2,11),wcss)
plt.title('Elbow Method')
plt.xlabel('No of Clusters')
plt.ylabel('WCSS/SSE')
plt.show()
```

##### As per Elbow method no of cluster should be 4

### Kneelocator Method

```python
knee=KneeLocator(range(2,11),wcss,curve='convex',direction='decreasing')
knee.elbow
```

##### As per Kneelocator method no of cluster should be 4

### Validatiing no of clusters using silhouette_coefficient
We have already calculated score while doing kmeans cluster identifiaction previously. plotting the graph

```python
plt.plot(range(2, 11), silhouette_coefficients)
plt.xlabel("Number of Clusters")
plt.ylabel("Silhouette Coefficient")
plt.show()
```

##### Silhoute Coefficient is giving cluster no 4 having highest score. so we will chose 4 as no of clusters

### Fitting Kmeans with no of cluster as 4

```python
kmeans=KMeans(n_clusters=4,init='k-means++',random_state=32)
y_kmeans=kmeans.fit(X)
cluster=kmeans.labels_
cluster
```

##### Kmeans have assigned data points to 4 clusters
